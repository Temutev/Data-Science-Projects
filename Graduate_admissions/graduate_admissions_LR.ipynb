```python
df.drop(['Serial No.'],axis=1,inplace=True)
df.head(2)
```

```python
g= sns.pairplot(df,height=6)
```

```python
#splitting  features from target 
X= df.drop(['Chance of Admit '],axis=1)
y= df['Chance of Admit ']
X.shape,y.shape
```

```python
#splitting our datasets into train and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
r = LinearRegression()
rfr = RandomForestRegressor(random_state=23)
etr =ExtraTreesRegressor()
dtr = DecisionTreeRegressor()
svrr = SVR()
gbr= GradientBoostingRegressor()
abr = AdaBoostRegressor()
ridge= BayesianRidge()
lasso = Lasso()

estimators =[
             ('LinearRegression',lr),
             ('RandomForestRegressor',rfr),
             ('ExtraTreesRegressor',etr),
             ('DecisionTreeRegressor',dtr),
             ('SVR',svrr),
             ('GradientBoostingRegressor',gbr),
             ('AdaBoostRegressor',abr),
            ('BayesianRidge',ridge),
            ('Lasso',lasso)
]
scores =[]
def model_training(X_train,X_test,y_train,y_test):
    
    for e,r in estimators:
        
        model_names = [e for e,r  in estimators]
        r.fit(X_train,y_train)
        r_preds = r.predict(X_test)
        scores.append(mae(y_test,r_preds))
        print("*****************************************************")
        print("Mean Absolute Error   for  {}  is :===>>>  ".format(e),mae(y_test,r_preds))
        print("*****************************************************")
        print("\n")

model_training(X_train,X_test,y_train,y_test)   
```

```python
#using linear regression
r =sns.regplot(x=y_test, y=lr.predict(X_test));
r.set_xlabel("Y Test")
r.set_ylabel("Y Predicted")
r.set_title('Relationship between the test variables and the predicted variables.')
```
