## d) Methodology/ML techinque used 

Linear Regression algorithm shows the relationship among variables and how the change in one independent variable impacts the other, called dependent variable. So, after I confirmed most of the variables are correlated, I decided to use linear regression because I want to predict the value of a variable (chance to be admitted) based on the value of others variables (GRA Score, TOEFL Score, etc). And, linear regression is one of the most interpretable machine learning algorithms, it is easy to explain, and it is easy of use.

### Linear Regression

```python
#Defining features (X) and target (y)
X = admissions[['GRE_Score', 'CGPA','TOEFL_Score']]
y = admissions['Chance_of_Admit_']
```

```python
#Splitting the data into training and testing sets
X_training, X_test, y_training, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

### Training the Model

```python
#Creating an instance of a LinearRegression() model named lm
lm = LinearRegression()
```

```python
#Training and fittin lm on the training data
lm.fit(X_training, y_training)
```

### Predicting Test Data

```python
y_predicted = lm.predict(X_test)
```

### Present the results both for the training and the validation dataset

```python
results = pd.DataFrame()
results['Test'] = y_test
results['Predictions'] = y_predicted

print(round(results[['Test','Predictions']].head(15),2))
```

```python
# Creating a scatterplot of  real test values vs predicted values.

plt.scatter(y_test,y_predicted)
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
```

## e) Measuring the Model

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

MAE = mean_absolute_error(y_test, y_predicted)
MSE = mean_squared_error(y_test, y_predicted)
RMSE = np.sqrt(MSE)

print('MAE: {}'.format(MAE))
print('MSE: {}'.format(MSE))
print('RMSE: {}'.format(RMSE))
```

```python
#Using a histogram of the residuals and make sure it looks normally distributed. 
sns.distplot(y_test-y_predicted)
```

```python
#Verifying the coefficients
coeficient = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeficient
```
